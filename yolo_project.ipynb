```python
import yaml

# 드라이브에 있는 yaml 파일 읽기
with open('/content/drive/MyDrive/yolo/data.yaml', 'r') as f:
    data = yaml.safe_load(f)

# 경로 추가해 드라이브 내 폴더 읽을수 있게 변경
data['path'] = '/content/drive/MyDrive/yolo'
data['train'] = 'train/images'
data['val'] = 'valid/images'
data['test'] = 'test/images'

# 수정된 내용 저장
with open('/content/drive/MyDrive/yolo/data.yaml', 'w') as f:
    yaml.dump(data, f)
```

```python
# train.py를 실행시켜 yolov5를 학습
!python train.py \
  --img 640 \
  --batch 16 \
  --epochs 200 \
  --data /content/drive/MyDrive/yolo/data.yaml \
  --weights yolov5s.pt \
  --project /content/drive/MyDrive/yolo_results \
  --hyp data/hyps/hyp.scratch-high.yaml \
  --cache
```

```python
# 테스트 용 이미지를 통해 학습된 모델 작동 확인
!python val.py \
  --weights /content/drive/MyDrive/yolo_results/exp/weights/best.pt \
  --data /content/drive/MyDrive/yolo/data.yaml \
  --img 640 \
  --batch 16 \
  --task test \
  --save-txt \
  --save-conf \
  --project /content/drive/MyDrive/yolo_results \
  --name test_results
```
